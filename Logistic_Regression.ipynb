```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("diabetes.csv")
```

```python
df.head()
```

```python
df=df.rename(columns={'Outcome':'Label'})
```

```python
df
```

```python
X=df.iloc[:,0:-1]
```

```python
Y=df.iloc[:,-1]
```

```python
X
```

```python
Standard=StandardScaler()
Standard=Standard.fit(X)
X=pd.DataFrame(Standard.fit_transform(X))
X
```

```python
Y
```

```python
X_train, X_test, Y_train, Y_test=train_test_split(X,Y, test_size=0.3)
```

```python
lgtest=LogisticRegression()
lgtest=lgtest.fit(X_train,Y_train)
```

```python
Y_pred=lgtest.predict(X_test)
```

```python
Y_pred
```

```python
from sklearn import metrics
cnf_matrix = metrics.confusion_matrix(Y_test, Y_pred)
cnf_matrix
```

```python
test=metrics.accuracy_score(Y_test,Y_pred)
test
```
