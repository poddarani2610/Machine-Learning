```python
#setting the half life for the independent media variables
Half_life_value={'imp_tv_cricket':2,'imp_tv_gec':0.005,'imp_tv_info':2,'imp_tv_movie':2,'imp_tv_news':2,'Paid_Brand_Mktg_Clicks':1}
#creating independent variable dataset
X=df_PBM[['imp_tv_cricket','imp_tv_gec','imp_tv_info','imp_tv_movie','imp_tv_news','Paid_Brand_Mktg_Clicks']].astype('float')
Y=df_PBM[['users_paid_brand_mktng']].astype('float')
#adstocking the variables
for i in X:
    for j in range(0,len(X)):
        if j==0:
            X[i][j]=X[i][j]
        else:
            X[i][j]=round(X[i][j]+(1/pow(2,Half_life_value[i]))*X[i][j-1],3)


  
#Gamma transformation of media variables
#setting effectiveness parameter
E={'imp_tv_cricket':2,'imp_tv_gec':2,'imp_tv_info':2,'imp_tv_movie':2,'imp_tv_news':1,'Paid_Brand_Mktg_Clicks':2}
#setting saturation parameter
S={'imp_tv_cricket':10,'imp_tv_gec':10,'imp_tv_info':10,'imp_tv_movie':10,'imp_tv_news':10,'Paid_Brand_Mktg_Clicks':10}

def hill_transform(x, ec, slope):
    return 1 / (1 + (x / ec)**(-slope))

for i in X:
    X[i]=hill_transform(X[i],ec=E[i],slope=S[i])
X
```

```python
#splitting the dataset into training and testing data
X_train, X_test, Y_train, Y_test=train_test_split(X,Y, test_size=0.3)
#standardizing the training data
std_scal=StandardScaler()
# std_scal=std_scal.fit(X)
X_train=pd.DataFrame(std_scal.fit_transform(X_train))
print(X_train)   
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
  
# VIF dataframe
vif_data = pd.DataFrame()
vif_data["feature"] = X_train.columns
  
# calculating VIF for each feature
vif_data["VIF"] = [variance_inflation_factor(X_train.values, i)
                          for i in range(len(X_train.columns))]
  
print(vif_data.sort_values(by='VIF'))
```
