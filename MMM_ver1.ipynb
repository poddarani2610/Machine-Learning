```python
#setting the half life for the independent media variables
Half_life_value={'imp_tv_cricket':1,'imp_tv_gec':1,'imp_tv_info':1,'imp_tv_movie':1,'imp_tv_news':1,'Paid_Brand_Mktg_Clicks':2}
#creating independent variable dataset
X=df_PBM[['imp_tv_cricket','imp_tv_gec','imp_tv_info','imp_tv_movie','imp_tv_news','Paid_Brand_Mktg_Clicks']].astype('float')
Y=df_PBM[['users_paid_brand_mktng']].astype('float')
#adstocking the variables
for i in X:
    for j in range(0,len(X)):
        if j==0:
            X[i][j]=X[i][j]
        else:
            X[i][j]=round(X[i][j]+(1/pow(2,Half_life_value[i]))*X[i][j-1],3)
        print(X[i][j])
```

```python
#splitting the dataset into training and testing data
X_train, X_test, Y_train, Y_test=train_test_split(X,Y, test_size=0.3)
```

```python
#importing the linear regression 
reg=LinearRegression()
```

```python
reg=reg.fit(X_train,Y_train)
```

```python
Y_pred=reg.predict(X_test)
```
